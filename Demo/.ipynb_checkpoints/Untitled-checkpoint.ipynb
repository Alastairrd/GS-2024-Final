```python
import numpy as np
import random

# Grid dimensions
rows, cols = 11, 11  # Must be odd for proper maze generation

# Initialize the grid
grid = np.ones((rows, cols), dtype=int)
grid[1::2, 1::2] = 0  # Cells are 0, walls are 1

# Union-Find structure for Kruskal's algorithm
class UnionFind:
    def __init__(self, n):
        self.parent = list(range(n))
        self.rank = [0] * n

    def find(self, x):
        if self.parent[x] != x:
            self.parent[x] = self.find(self.parent[x])  # Path compression
        return self.parent[x]

    def union(self, x, y):
        rootX = self.find(x)
        rootY = self.find(y)
        if rootX != rootY:
            if self.rank[rootX] > self.rank[rootY]:
                self.parent[rootY] = rootX
            elif self.rank[rootX] < self.rank[rootY]:
                self.parent[rootX] = rootY
            else:
                self.parent[rootY] = rootX
                self.rank[rootX] += 1

# Create an edge list
edges = []
for i in range(1, rows, 2):
    for j in range(1, cols, 2):
        # Add edges for adjacent cells
        if i + 2 < rows:
            edges.append(((i, j), (i + 2, j)))  # Vertical wall
        if j + 2 < cols:
            edges.append(((i, j), (i, j + 2)))  # Horizontal wall

# Shuffle edges using Fisher-Yates algorithm
random.shuffle(edges)

# Initialize Union-Find
cell_count = (rows // 2) * (cols // 2)
uf = UnionFind(cell_count)
print(uf.parent)
print(uf.rank)
# Map cells to IDs
cell_to_id = {(i, j): (i // 2) * (cols // 2) + (j // 2) for i in range(1, rows, 2) for j in range(1, cols, 2)}

# Build a perfect maze
for edge in edges:
    cell1, cell2 = edge
    id1 = cell_to_id[cell1]
    id2 = cell_to_id[cell2]
    if uf.find(id1) != uf.find(id2):
        # Remove the wall between cell1 and cell2
        wall = ((cell1[0] + cell2[0]) // 2, (cell1[1] + cell2[1]) // 2)
        grid[wall] = 0
        uf.union(id1, id2)

# Add cycles based on cycle bias
cycle_bias = 0.2  # 20% openness
num_cycles = int(cycle_bias * (rows * cols // 2))

remaining_edges = [edge for edge in edges if grid[((edge[0][0] + edge[1][0]) // 2, (edge[0][1] + edge[1][1]) // 2)] == 1]
random.shuffle(remaining_edges)

for _ in range(min(num_cycles, len(remaining_edges))):
    edge = remaining_edges.pop()
    wall = ((edge[0][0] + edge[1][0]) // 2, (edge[0][1] + edge[1][1]) // 2)
    grid[wall] = 0

# Introduce horizontal/vertical bias
horizontal_bias = 0.7  # 70% horizontal paths
vertical_bias = 1 - horizontal_bias

horizontal_walls = [edge for edge in edges if edge[0][0] == edge[1][0]]
vertical_walls = [edge for edge in edges if edge[0][1] == edge[1][1]]

num_horizontal = int(horizontal_bias * len(horizontal_walls))
num_vertical = int(vertical_bias * len(vertical_walls))

for wall_set, target_count in [(horizontal_walls, num_horizontal), (vertical_walls, num_vertical)]:
    random.shuffle(wall_set)
    for edge in wall_set[:target_count]:
        wall = ((edge[0][0] + edge[1][0]) // 2, (edge[0][1] + edge[1][1]) // 2)
        if grid[wall] == 1:
            grid[wall] = 0

# Print the resulting maze
print("Generated maze:")
print(grid)
```

```python
class Node:
    
    def __init__(self, row, col):
        self.row = row
        self.col = col
        self.neighbours=[]
        self.is_wall = True

class Map:
    
    def __init__(self, map_size):
        self.grid = [[Node(row, col) for col in range(map_size)] for row in range(map_size)]
        self.size = map_size

class UnionFind:
    
    def __init__(self, map):
        self.parent=[[(r, c) for c in range(map.size)] for r in range(map.size)] #parents are coordinates as id
        self.rank=[[[0] for c in range(map.size)] for r in range(map.size)] #rank for union by rank
        print("rank:")
        print(self.rank)
        
    def find(self, coord):
        r, c = coord
        if self.parent[r][c] != (r,c):
            self.parent[r][c] = self.find(self.parent[r][c]) #path compression
        return self.parent[r][c]
    
    def union(self, coord1, coord2):
        root1 = self.find(coord1)
        root2 = self.find(coord2)
        
        if root1 != root2:
            r1, c1 = coord1
            r2, c2 = coord2
            
            #check ranks and attach lower rank tree to the other 
            if self.rank[r1][c1] > self.rank[r2][c2]:
                self.parent[r2][c2] = root1
            elif self.rank[r1][c1] < self.rank[r2][c2]:
                self.parent[r1][c1] = root2
            else: ##if same just attach tree 2 to first
                self.parent[r2][c2] = root1
                print("rank[r1][c1]")
                print(self.rank[r1][c1])
                self.rank[r1][c1][0] += 1
                
#init the map with size
map = Map(15)
#iterate through map rows and cols
for row in range(map.size):
    for col in range(map.size):
        node = map.grid[row][col]
        if row > 0:
            node.neighbours.append(map.grid[row - 1][col]) #north neighbour
        if row < map.size - 1:
            node.neighbours.append(map.grid[row + 1][col])  #south neighbour
        if col > 0:
            node.neighbours.append(map.grid[row][col - 1])  #west neighbour
        if col < map.size - 1:
            node.neighbours.append(map.grid[row][col + 1])  #east neighbour

#set initial path cells
for row in range(map.size):
    for col in range(map.size):
        if row % 2 == 1 and col % 2 == 1:
            map.grid[row][col].is_wall = False
            
def drawMap(map):
    for row in range(map.size):
            print(''.join('0' if not node.is_wall else '#' for node in map.grid[row]))

drawMap(map)
print(" ")


```

```python
import random

# Step 1: Create the Edge List
edges = []
for row in range(1, map.size, 2):  # Only iterate over path cells
    for col in range(1, map.size, 2):
        if row + 2 < map.size:  # Vertical wall
            wall = map.grid[row + 1][col]
            cell1 = map.grid[row][col]
            cell2 = map.grid[row + 2][col]
            edges.append((wall, cell1, cell2))  # (wall, cell1, cell2)
        if col + 2 < map.size:  # Horizontal wall
            wall = map.grid[row][col + 1]
            cell1 = map.grid[row][col]
            cell2 = map.grid[row][col + 2]
            edges.append((wall, cell1, cell2))

# Step 2: Shuffle the Edge List
random.shuffle(edges)

# Step 3: Apply Kruskal's Algorithm
uf = UnionFind(map)  # Initialize Union-Find with the map

for wall, cell1, cell2 in edges:
    coord1 = (cell1.row, cell1.col)
    coord2 = (cell2.row, cell2.col)
    
    if uf.find(coord1) != uf.find(coord2):  # If not in the same set
        uf.union(coord1, coord2)  # Union the two cells
        wall.is_wall = False  # Remove the wall between the cells

# Step 4: Draw the Final Maze
drawMap(map)
```

```python
# Generate edges between cells
edges = []
cell_to_id = {}  # Map cells to unique IDs
counter = 0

# Assign unique IDs to cells and add edges
for row in range(1, map.size, 2):
    for col in range(1, map.size, 2):
        node = map.grid[row][col]
        cell_to_id[(row, col)] = counter
        counter += 1
        # Vertical neighbor
        if row + 2 < map.size:
            neighbor = map.grid[row + 2][col]
            edges.append((node, neighbor))
        # Horizontal neighbor
        if col + 2 < map.size:
            neighbor = map.grid[row][col + 2]
            edges.append((node, neighbor))
```

```python
# Initialize Union-Find
uf = UnionFind(counter)

# Shuffle edges for randomness
random.shuffle(edges)

# Build the maze (Perfect Maze)
for edge in edges:
    node1, node2 = edge
    id1 = cell_to_id[(node1.row, node1.col)]
    id2 = cell_to_id[(node2.row, node2.col)]

    # Check if nodes are in different sets
    if uf.find(id1) != uf.find(id2):
        # Union the sets
        uf.union(id1, id2)

        # Remove the wall between the two cells
        wall_row = (node1.row + node2.row) // 2
        wall_col = (node1.col + node2.col) // 2
        grid[wall_row][wall_col].is_wall = False  # Remove wall
        
# Add cycles for a non-perfect maze
cycle_bias = 0.2  # Fraction of edges to add back as cycles
remaining_edges = [edge for edge in edges if uf.find(cell_to_id[(edge[0].row, edge[0].col)]) == uf.find(cell_to_id[(edge[1].row, edge[1].col)])]
random.shuffle(remaining_edges)

num_cycles = int(cycle_bias * len(edges))
for i in range(min(num_cycles, len(remaining_edges))):
    node1, node2 = remaining_edges[i]
    wall_row = (node1.row + node2.row) // 2
    wall_col = (node1.col + node2.col) // 2
    grid[wall_row][wall_col].is_wall = False  # Remove wall
    
# Print the maze
for row in grid:
    print(''.join(' ' if not node.is_wall else '#' for node in row))
```
